```python
import time
from turtle import Screen
from player import Player
from car_manager import CarManager
from scoreboard import Scoreboard

screen = Screen()
screen.setup(width=600, height=600)
screen.tracer(0)

player = Player()
board = Scoreboard()
car = CarManager()

screen.listen()
screen.onkeypress(fun=player.up, key='Up')

game_is_on = True
while game_is_on:
    time.sleep(0.1)
    screen.update()

    car.create_car()
    car.move()

    # Detect if turtle crossed the screen
    if player.ycor() > 280:
        board.level_up()
        player.new_level()
        car.car_speedup()

    # Detect collision with the paddle
    for i in car.all_cars:
        if player.distance(i) < 20:
            board.game_over()
            game_is_on = False  

screen.exitonclick()
        
```
