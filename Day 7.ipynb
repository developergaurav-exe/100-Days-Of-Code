### Hangman Project

```python
from hangman_words import word_list
stages = [

'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''',

'''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', 

'''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', 

'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========
''', 

'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', 

'''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', 

'''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

logo = ''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    '''
```

```python
import random
lives = 6
chosen_word = random.choice(word_list)

display = ['_' for x in range(len(chosen_word))]
print(logo)
print(f"\n{' '.join(display)}")
game_over = False
# print(f'Pssst, the solution is {chosen_word}.') This is a testing Code do not cheat :p
while not game_over:
    guess = input('Guess a letter: ').lower()

    if guess in display:
        print(f'You have already guessed the Word {guess}')

    for i in range(len(chosen_word)):
        if guess == chosen_word[i]:
            display[i] = guess
        
    if guess not in chosen_word: 
        lives -= 1
        print(f'{guess} is not in the word. You lose a life, Remaining lives {lives}')
        if lives == 0:
            print(f'You Lose. Answer is {chosen_word}')
            game_over = True

    print(stages[lives])
    print(f"{' '.join(display)}")            

    if '_' not in display:
        game_over = True
        print('You Win')
```
